```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
```

```python
page
```

```python
top = BeautifulSoup(page.content)
```

```python
name = top.find("td",class_='titleColumn').text.strip().replace('1.\n','').replace('\n','').replace('(1994)','')
name
```

```python
name_ = []
for name in top.find_all("td",class_='titleColumn'):
    name_.append(name.text.strip().replace('\n','').replace('\n','').replace('()',''))
name_    
```

```python
rating = top.find("td", class_= 'ratingColumn imdbRating')
rating.text.strip()
```

```python
rating_ = []
for r in top.find_all("td", class_= 'ratingColumn imdbRating'):
    rating_.append(r.text.strip())
rating_
```

```python
year = top.find("span", class_="secondaryInfo")
```

```python
year.text
```

```python
year_ = []
for year in top.find_all("span", class_="secondaryInfo"):
    year_.append(year.text)
year_
```
